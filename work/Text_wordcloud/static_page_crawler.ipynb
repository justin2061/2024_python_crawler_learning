### 使用 requests.get() 方法取得資料

```python
url = 'https://udn.com/news/story/123777/8266151'
```

```python
import requests

res = requests.get(url)
print(f"status code: {res.status_code}")
print(f"response headers: {res.headers}")
print(f"response body: {res.text}")
```

### 使用 requests.post() 方法取得資料

```python
#- 來源：MOI_API_039公有土地開放資料服務 https://cop.land.moi.gov.tw/Services/service_interfaceInfo.aspx?da8a69fd-673a-4a56-9548-4b11dfbf02e2
import requests

url = "https://openapi.land.moi.gov.tw/WEBAPI/PublicLandOData/QueryByTown"

payload = "[{\"CITY\":\"A\",\"TOWN\":\"03\",\"SEC\":\"0001\"}]"

headers = {
    'accept': "application/json",
    'Content-Type': "application/json"
    }

response = requests.request("POST", url, data=payload, headers=headers)
if response.status_code == 200:
    print(response.json())
```

```python
import pandas as pd
data = response.json()['RESPONSE'][0]['NATIONAL']['PUBLIC']
df = pd.DataFrame(data)
df
```
